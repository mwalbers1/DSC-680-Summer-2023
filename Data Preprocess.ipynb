```python
df_cardinality = pd.DataFrame( { 'categorical variable': ['app', 'channel', 'device', 'os'],
                                'cardinality': [
                                df['app'].nunique(),
                                df['channel'].nunique(),
                                df['device'].nunique(),
                                df['os'].nunique() ]
                              }).T
```

```python
df_cardinality.reset_index(inplace=True)
```
